### Analysing data

```python
df.describe().T
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(x=df['YEAR'],y=df[' ANNUAL RAINFALL'],marker='o',color='indigo')
plt.title('Annual rainfall')
plt.show()
```

### Kerala had the highest rainfall in 2018

```python
a=df[df['YEAR']==2018].squeeze()
months=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
plt.figure(figsize=(10,5))
plt.bar(months,a[months],color='violet',edgecolor='black')
plt.title('rainfall of each month in 2018')
plt.xlabel('months')
plt.ylabel('rainfall recieved')
plt.show()
```

### In 2018 Kerala recieved the highest rainfall in the month of august

```python
plt.figure(figsize=(15,7))
df['FLOODS'].value_counts().plot(kind='pie',colors=['gold','silver'],autopct='%1.1f%%',startangle=90)
plt.title('flood occured')
plt.show()
```

### From 1901 to 2018 50% of time kerala had flood

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, hue='FLOODS',y=' ANNUAL RAINFALL',palette='inferno')
plt.title('rainfall recieved annually during flood and non flood years')
plt.show()
```

### When there is an annual rainfall >3300 ,its more likely to have a flood

```python
monthly_avg = df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 
                  'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].mean()
plt.figure(figsize=(14, 6))
sns.barplot(x=monthly_avg.index, y=monthly_avg.values,color="navy")
plt.title('Average Monthly Rainfall in Kerala (1901â€“2018)', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Average Rainfall (mm)')
plt.show()
```

### Data indicates that the july month receive the highest rainfall

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='YEAR', y='JUN', color='teal', marker='o')
plt.title('July Rainfall Trend Over the Years in Kerala', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Rainfall in June (mm)')
plt.grid(True)
plt.show()
```

### july month recieved highest rainfall in 1990

```python
df['Decade'] = (df['YEAR'] // 10) * 10
plt.figure(figsize=(12, 6))
sns.countplot(x='Decade', hue='FLOODS', data=df, palette='inferno')
plt.title('Flood Occurrences by Decade', fontsize=16)
plt.xlabel('Decade')
plt.ylabel('Count')
plt.legend(title='Floods')
plt.show()
```

### 1940 to 1950 decade had the highest flood years and 1980 to 1990 had the least flood years

```python
df['Monsoon_Rainfall'] = df[['JUN', 'JUL', 'AUG', 'SEP']].sum(axis=1)
df['Non_Monsoon_Rainfall'] = df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'OCT', 'NOV', 'DEC']].sum(axis=1)
```

```python
plt.figure(figsize=(30,10))
df[['YEAR', 'Monsoon_Rainfall', 'Non_Monsoon_Rainfall']].set_index('YEAR').plot(kind='line')
plt.title('Monsoon vs Non-Monsoon Rainfall in Kerala', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()
```

### Kerala depends upon the monsoon months for rain as it recieves highest rainfall during monsoon
